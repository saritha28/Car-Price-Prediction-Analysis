### Findinga and Removing Outliers

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#Finding outliers
numeric_cols= df.select_dtypes(include=['int64','float64'])
plt.figure(figsize=(18,10))
sns.boxplot(data=numeric_cols)
plt.show()
```

We can find outliers in the price column.We need to work on outliers.

```python
def remove_outliers(cols):
    sorted(cols)
    Q1,Q3 = cols.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
for i in numeric_cols.columns:
    lower_range,upper_range = remove_outliers(df[i])
    df[i] = np.where(df[i] > upper_range,upper_range,df[i])
    df[i] = np.where(df[i] < lower_range,lower_range,df[i])
```

```python
#Again finding outliers
numeric_cols = df.select_dtypes(include=(['int64','float64']))
plt.figure(figsize=(20,10))
sns.boxplot(data=numeric_cols)
plt.show()
```

Now we can see that there are no outliers in the numeric columns.

# Data Splitting into Train and Test 

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("Y_train:", Y_train.shape)
print("Y_test:", Y_test.shape)
```

### Standardize the data

```python
scale = StandardScaler()
X_train_scale = scale.fit_transform(X_train)
X_test_scale = scale.fit_transform(X_test)
```

```python
X_train_scale
```

## PCA

```python
pc = PCA()
X_train_scale_pc = pc.fit_transform(X_train_scale)
X_test_scale_pc = pc.fit_transform(X_test_scale)
```

```python
#Explained variance ratio 
#The explained variance ratio is an array of the variance of the 
#data explained by each of the principal components in your data. 
exp_var = pc.explained_variance_ratio_
```

```python
print("Explained Variance Ratio:", exp_var)
```

```python
#Cumulative sum of eigenvalues; This will be used to create step plot
#for visualizing the variance explained by each principal component.

#Now let us calculate the cumulative sum of explained variance ratio
cum_var = np.cumsum(exp_var)
```

```python
#plot the scree plot or cumulative explained variance plot
plt.plot(range(1,len(exp_var)+1),cum_var,marker='o',linestyle='-')
plt.xlabel('Number of COmponents')
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree plot or Cumulative Explained Variance Plot")
plt.grid(True)
plt.show()
```

```python
#cumulative explained variance 
cum_var = np.cumsum(exp_var)

#how many components to select 
#here we are capturing 95% variance
desired_var = 0.95

#how many components ca we select
num_comp = np.argmax(cum_var >= desired_var)+1

print(f"\number of components selected:{num_comp}")
```

```python
pc = PCA(n_components=12)
X_train_scale_pc = pc.fit_transform(X_train_scale)
X_test_scale_pc = pc.fit_transform(X_test_scale)
```

```python
exp_var = pc.explained_variance_ratio_
print("Explained Variance Ratio:",exp_var)
```

```python
#Now let us calculate the cumulative sum of explaine variance ratio
cum_var = np.cumsum(exp_var)
```

```python
#Let us plot the scree plot or cumulative explained variance plot
plt.plot(range(1,len(exp_var)+1),cum_var,marker='o',linestyle='-')
plt.xlabel('Number of components')
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree plot or Cumulative Explained Variance plot")
plt.grid(True)
plt.show()
```
